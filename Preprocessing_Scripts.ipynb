```python
def calculate_average(range_value):
    # Remove 'Years' from the string and split by ' to '
    range_value = range_value.replace(' Years', '')
    numbers = list(map(int, range_value.split(' to ')))  # Convert parts to integers
    return sum(numbers) / len(numbers)  # Calculate average

# Apply the function to the 'Experience' column and create a new column
df['Average_Experience'] = df['Experience'].apply(calculate_average)

print(df.Average_Experience)
```

```python
df.Salary_Range
```

```python
def calculate_salary_average(salary_range):
    # Remove the dollar sign ($) and 'K', then split by '-'
    salary_range = salary_range.replace('$', '').replace('K', '')
    numbers = list(map(int, salary_range.split('-')))  # Convert parts to integers
    return sum(numbers) / len(numbers)  # Calculate average

# Apply the function to the 'Salary Range' column and create a new column
df['Average_Salary'] = df['Salary_Range'].apply(calculate_salary_average)

print(df.Average_Salary)
```
